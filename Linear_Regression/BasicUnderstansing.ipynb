```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#basic data manually created
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
```

```python
# this is a rough guess, we will get actuall value below by formulae in scikit learn
m=2
b=0
```

```python
y_pred = m*x + b #formulae
```

```python
plt.scatter(x,y)
plt.plot(x, y_pred)
plt.show()
```

```python
error = y - y_pred #formulae of error
error
```

```python
# use in scikit library
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model = model.fit(x.reshape(-1,1), y)
```

```python
# getting value of m
print(model.coef_)
print(f"which means: For every +1 in x, y increases by {model.coef_[0]}.")
```

```python
# getting value of b
print(model.intercept_)
print("Meaning: When x = 0, y = 0.")
```
